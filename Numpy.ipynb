# Numpy Arrays

```python
import numpy as np
```

```python
oneDArray = np.array([1,2,3,4,5])
print(oneDArray)
```

```python
print(oneDArray.ndim) #dimension
```

```python
twoDArray = np.array([[1,2,3], [5,6,7]])
print(twoDArray)
```

```python
print(twoDArray.ndim)
```

```python
threeDArray = np.array([[[1,2,3], [5,6,7]], [[2,7,8], [-9,-8,-6]]])
print(threeDArray)
```

```python
print(threeDArray.ndim)
```

```python
twoDArray = np.array([[1,2,3], [5,6,7]])
print(twoDArray.shape) #rows and columns
```

# Zeros

```python
zerosArray = np.zeros(10)
print(zerosArray)
```

```python
zerosArray = zerosArray.astype(int)
print(zerosArray)
```

# Ones

```python
onesArray = np.ones(10)
print(onesArray)
```

```python
onesArray = onesArray.astype(int)
print(onesArray)
```

# Full

```python
preFilledArray = np.full(10,5)
print(preFilledArray)
```

# Statistics

```python
matrix1 = np.array([[1,2,3], [4,5,6]])
print(matrix1)
```

```python
np.min(matrix1)
```

```python
np.max(matrix1)
```

```python
np.sum(matrix1)
```

```python
np.mean(matrix1)
```

```python
np.std(matrix1)
```

```python
np.median(matrix1)
```
