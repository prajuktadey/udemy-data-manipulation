# Dictionaries

```python
data  = {"Name": "Prajukta", "Age": 30, "Country": "USA"}
name = data["Name"]
age = data["Age"]
country = data["Country"]

print(name)
print(age)
print(country)
```

```python
data  = {"Name": "Prajukta", "Age": 30, "Country": "USA"}
for key in data:
  print (data[key])
```

# User Defined Functions

```python
def AddNumber(x,y):
  sum = x+y
  return sum

result = AddNumber(5,7)
print(result)
```

```python
def IncrementsByOne(x):
  increment= x+1
  return increment

result= IncrementsByOne(7)
print(result)
```
