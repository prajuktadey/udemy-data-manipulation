# Title, Labels & Legend

```python
x_axis = [1,2,3]
y_axis = [4,5,6]

plt.plot(x_axis, y_axis, 'r')
plt.title('My first Graph')
```

```python
x_axis = [1,2,3]
y_axis = [4,5,6]

plt.plot(x_axis, y_axis, 'r')
plt.title('My first Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

# Legends

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10, 100, 1000]

x2_axis = [1,3,5,7]
y2_axis = [100, 110, 120, 130]

plt.plot(x1_axis, y1_axis, 'r')
plt.plot(x2_axis, y2_axis, 'g')

plt.title('Two Plots on One Chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend(['exponential', 'linear'])
```

# Plotting Pie Charts

# Plotting Polar Plots

```python
import numpy as np
```

```python
theta = np.arange(0,(2*np.pi), 0.01) #generating an array of evenly spaced floats
r=2

for radian in theta:
  plt.polar(radian, r)
```

# Creating multiple subplots in one figure

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,100]

x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]

plt.subplot(1,2,1)
plt.plot(x1_axis, y1_axis, 'r')

plt.subplot(1,2,2)
plt.plot(x2_axis, y2_axis, 'g')

plt.tight_layout()
```
